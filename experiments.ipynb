### 6.1 k-Means CLUSTERING METHOD

```python
from sklearn.cluster import KMeans 
from sklearn.metrics import silhouette_score
from yellowbrick.cluster import SilhouetteVisualizer
from warnings import filterwarnings
filterwarnings('ignore')
```

#### 6.1.1  Silhouette Score

```python
def plot_silhouette_scores(X, upper_limit):
    silhouettes = []
    for k in range(2, upper_limit):
        kmeans = KMeans(n_clusters=k, random_state = 1, init = 'k-means++', n_init = 30)
        cluster_assignments = kmeans.fit_predict(X)
        silhouettes.append(silhouette_score(X, cluster_assignments))
        plt.plot(range(2, k+1), silhouettes)
        plt.xlabel("Clusters")
        plt.ylabel("silhouette_score")
plot_silhouette_scores(X, 10)
```

<div class="alert alert-block alert-warning">
<b>n_cluster =</b>

###### The number of `clusters=4` has the maximum score based on the silhouette score plot above. Therefore, I set the **k = 4** for k-Means clustering.


#### 6.1.2  Silhouette Plot

```python
def KMeansParameters(n_clusters=2, random_state=100, n_init=20, max_iter=300, tol=0.0001): 
    return KMeans(n_clusters=n_clusters, random_state=random_state, n_init=n_init, init='k-means++', max_iter=max_iter, tol=tol)

colours=["red", "blue", "yellow", "purple"]
kmeans = KMeansParameters(n_clusters=4)
cluster_assignments_kmeans=kmeans.fit_predict(X)

visualizer_B = SilhouetteVisualizer(kmeans, is_fitted = True, colors=colours)
visualizer_B.fit(X)
visualizer_B.finalize()
```

### 6.2 HAC CLUSTERING METHOD

#### 6.2.1  Dendogram

```python
from sklearn import cluster
from scipy.cluster.hierarchy import dendrogram
from scipy.cluster import hierarchy
```

```python
agg = cluster.AgglomerativeClustering(
    n_clusters=4, 
    metric='euclidean', 
    linkage='complete', 
    compute_distances=True)
labels = agg.fit_predict(X)
```

```python
def create_linkage(agg):
    n_samples = len(agg.labels_)
    counts = np.zeros(agg.children_.shape[0])
    for i, merge in enumerate(agg.children_):
        current_count = 0
        for child_idx in merge:
            if child_idx < n_samples:
                current_count += 1  # leaf node
            else:
                current_count += counts[child_idx - n_samples]
        counts[i] = current_count
    linkage_matrix = np.column_stack([agg.children_, agg.distances_,counts]).astype(float)
    return linkage_matrix

matrix=create_linkage(agg)
```

```python
hierarchy.set_link_color_palette(['red', 'green', 'blue', 'purple'])
plt.figure(figsize=(18,5))
dendrogram(
    matrix,
    color_threshold=10, 
    orientation='top', 
    labels=X.index, 
    leaf_rotation=90)
plt.show();
```

###### I generated a dendogram, which visually represents the hierarchy of clusters and the distances at which they are merged. I decided to align the hierarchical clustering results with the k-means baseline to be able to make the best choice between two cluster methods, I identified threshold as 10 on the dendrogram corresponding to a similar number of clusters (in this case, 4 clusters).
